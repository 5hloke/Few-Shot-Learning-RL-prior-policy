```python
import Data_Prep
import train_Reptile
import train_MAML
import train_iMAML
```

```python
data = Data_Prep.DataPreparation()

ml10 = data.get_data()
data.construct_episodes(ml10)
X, y, task_lengths = data.prepare_data()
```

```python
import torch
input_size = 43
hidden_size1 = 256
hidden_size2 = 256
hidden_size3 = 256

imaml_model = train_iMAML.iMAML(input_size, hidden_size1, hidden_size2, hidden_size3)
imaml_train_x, imaml_train_loss, imaml_train_acc = imaml_model.train(X, y, task_lengths, 1000)
#save the model in the models directory

```

```python
torch.save(imaml_model.model.state_dict(), '../models/imaml_model.pth')
```

```python
# PEBBLE test with MAML
from pebble import Pebble
# get open drawer data
vdata = Data_Prep.DataPreparation()
valid = vdata.get_data(validation= True, task = 'cls_drawer-open', prior = True)
vdata.construct_episodes(valid)
X, y, task_lengths = vdata.prepare_data()
```

```python
model = train_iMAML.iMAML(input_size, hidden_size1, hidden_size2, hidden_size3)
model.model.load_state_dict(torch.load('../models/imaml_model.pth'))
pebble = Pebble(model)
imaml_drawer_open_x, imaml_drawer_open_y = pebble.train(X, y, task_lengths, 75)
```
