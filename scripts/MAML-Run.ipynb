```python
import Data_Prep
import train_MAML
import train_MAML
import train_iMAML
from pebble import Pebble

import torch
```

```python
%load_ext autoreload
%autoreload 2
```

```python
data = Data_Prep.DataPreparation()

ml10 = data.get_data()
data.construct_episodes(ml10)
X, y, task_lengths = data.prepare_data()
```

```python
input_size = 43
hidden_size1 = 256
hidden_size2 = 256
hidden_size3 = 256

MAML_model = train_MAML.MAML(input_size, hidden_size1, hidden_size2, hidden_size3)
maml_train_pltx, maml_train_plty, maml_train_acc = MAML_model.train(X, y, task_lengths, 1001)
```

### Save model

```python
torch.save(MAML_model.model.state_dict(), '../models/MAML_model.pth')
```
