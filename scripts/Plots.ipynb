```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
task_x = np.loadtxt("results/imaml/imaml_drawer_open_x.txt", delimiter=',')
imaml_drawer_open_prior_acc = np.loadtxt("results/imaml/imaml_drawer_open_y.txt")
maml_drawer_open_prior_acc = np.loadtxt("results/MAML/maml_valid_prior_draweropen_plty.txt")
reptile_drawer_open_prior_acc = np.loadtxt("results/reptile/rep_valid_prior_draweropen_plty.txt")
```

```python
plt.plot(task_x, imaml_drawer_open_prior_acc, label = "iMAML")
plt.plot(task_x, maml_drawer_open_prior_acc[:75], label = "MAML")
plt.plot(task_x, reptile_drawer_open_prior_acc[:75], label = "Reptile")
plt.axhline(0.9, linestyle='--', color='r')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Drawer-Open with prior")
plt.legend(loc = 'lower right')
plt.savefig('results/plots/drawer_open_prior.png')
plt.show()
```
